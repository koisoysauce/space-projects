```python
# 1. Which planet has the most moons?
df1 = df.copy().sort_values(by='number_of_moons', ascending=False)
df1.iloc[0]

# Jupiter has the most moons in our solar system
```

```python
# 2. Is there a relationship between planet size and number of moons?
df2 = df.copy()[['diameter', 'number_of_moons']].sort_values(by='diameter')
df2
```

```python
# 2. Plot diameter vs. moon count
diameters = np.array(df2['diameter'].to_list())
number_of_moons = np.array(df2['number_of_moons'].to_list())

plt.plot(diameters, number_of_moons, 'bo-')
plt.title('Moon Count vs. Planet Diameter (km)')
plt.xlabel('diameter (km)')
plt.ylabel('moon count')
plt.tight_layout()
plt.show()
print(f"correlation coefficient: {df2['diameter'].corr(df2['number_of_moons']):.4f}")

# There appears to be a positive linear correlation between the number of moons and planet diameter 
# with a correlation coefficient of 0.987
# However, due to the small sample size of n=9, we have no evidence of a relation between moon counts and planet diameter
```

```python
# 3. How does gravity compare across planets?
df3 = df.copy()
df3['M/r^2'] = (df3['mass'] * 1e24) / (df3['diameter'] / 2 * 1000) ** 2
df3 = df3.sort_values(by='M/r^2', ascending=True)
df3.head(9)
# The g of planets is calculated through Newton's Law of Universal Gravitation
# $g=\frac{G*M}{r^2}$ using the radius (r) and mass (M) of earth with the gravitational constant G
# Thus, g is directly related to the mass and inversely proportional to the squared radius of te planet.
# We can observe this relation by looking at some graphs
```

```python
# Make a graph of (M/r^2) vs. gravity
x_var = np.array(df3['M/r^2'])
y_var = np.array(df3['gravity'])
plt.plot(x_var, y_var, 'bo-')
plt.title("M/r² vs. Acceleration of Gravity")
plt.xlabel("M/r² (kg/m²)")
plt.ylabel("Acceleration of Gravity (m/s²)")
plt.tight_layout()

for _, row in df3.iterrows():
    if row['planet'] == 'Saturn':
        plt.annotate(row['planet'],
                     xy=(row['M/r^2'], row['gravity']),
                     textcoords="offset points",
                     xytext=(5, 0),
                     fontsize=8,
                     color='red')

plt.show()
# Slope should be G (6.674e-11)
G = (23.1-0.7) / (3.71348193e+11-1.03971942e+10)
print(f'correlation coefficient: {df3['M/r^2'].corr(df3['gravity']):.4f}')
print(f'Slope using first and last points: {G:.3e}') # 6.206e-11
print(f'Percent error from gravitational constant: {abs((G - 6.674e-11) / 6.674e-11 * 100):.2f}%') # 7.01%
print("We have confirmed Newton's law holds across real NASA data of planets.")
```

```python
# 4. How long is a year on each planet relative to Earth?
df4 = df.copy()
df4['orbital_period_in_earth_years'] = df4['orbital_period'] / 365.2
df4[['planet', 'orbital_period_in_earth_years']]
```

```python
# Make a bar chart of each planet's orbital period in earth years
plt.bar(df4['planet'], df4['orbital_period_in_earth_years'])
plt.xlabel('Planet')
plt.ylabel('Earth Years')
plt.title('Orbital Period Relative to Earth')
plt.tight_layout()
plt.show()
```

```python
# According to Kepler's Third Law, a planet's orbital period in earth years is proportional as shown
# $T^2 \propto R^3$ where T is the orbital period and R is the distance from the planet to the sun
# We can confirm this by making a T^2 vs. R^3 graph
df4 = df4.sort_values(by='distance_from_sun')
plt.plot(df4['distance_from_sun'] ** 3, df4['orbital_period_in_earth_years'] ** 2, 'bo-')
plt.title('T² (Orbital Period) vs. R³ (Distance from Sun)')
plt.xlabel('R³ (e6 km)')
plt.ylabel('T² (Earth years)')
plt.tight_layout()
plt.show()
# This gives us a linear graph, proving Kepler's Third Law
print(f'correlation coefficient: {(df4['distance_from_sun'] ** 3).corr(df4['orbital_period_in_earth_years'] ** 2):.4f}')
print("We see a linear graph formed from this graph, confirming Kepler's Third Law:", end=' ')
print("T² ∝ R³")
```

```python
# 5. How much do I weigh in each planet?
df5 = df.copy()
my_mass_kg = 55
df5['my_weight (N)'] = df5['gravity'] * my_mass_kg
df5[['planet', 'my_weight (N)']]
```

```python
# Weight is described as such by Newton's Second Law, F=mg
# Thus, we can make a gravity vs weight graph to confirm Newton's Second Law, where the slope must be my mass
df5 = df5.sort_values(by='my_weight (N)')
plt.plot(df5['my_weight (N)'], df5['gravity'], 'bo-')
plt.title('gravity (m/s²) vs. weight (N)')
plt.xlabel('weight (N)')
plt.ylabel('gravity (m/s²)')
plt.tight_layout()
plt.show()
m = (1270.5 - 38.5) / (23.1 - 0.7)
print(f"correlation coefficient: {df5['my_weight (N)'].corr(df5['gravity']):.4f}")
print(f'Slope using first and last points: {m:.2f}')
print("Percent error = 0.00%")
print("This linear graph confirms Newton's Second Law, which states F = mg.")
```
