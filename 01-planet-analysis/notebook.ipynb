```python
# 1. Which planet has the most moons?
df1 = df.sort_values(by='number_of_moons', ascending=False)
df1.iloc[0]

# Jupiter has the most moons in our solar system
```

```python
# 2. Is there a relationship between planet size and number of moons?
df2 = df[['diameter', 'number_of_moons']].sort_values(by='diameter')
df2
```

```python
# 2. Plot diameter vs. moon count
diameters = np.array(df2['diameter'].to_list())
number_of_moons = np.array(df2['number_of_moons'].to_list())

plt.plot(diameters, number_of_moons, 'bo-')
plt.title('Moon Count vs. Planet Diameter (km)')
plt.xlabel('diameter (km)')
plt.ylabel('moon count')
plt.show()
df2['diameter'].corr(df2['number_of_moons'])

# There appears to be a positive linear correlation between the number of moons and planet diameter 
# with a correlation coefficient of 0.987
# However, due to the small sample size of n=9, we have no evidence of a relation between moon counts and planet diameter
```

```python
# 3. How does gravity compare across planets?
df3 = df.copy()
df3['M/r^2'] = (df3['mass'] * 1e24) / (df3['diameter'] / 2 * 1000) ** 2
df3 = df.sort_values(by='M/r^2', ascending=True)
df3.head(9)
# The g of planets is calculated through Newton's Law of Universal Gravitation
# $g=\frac{G*M}{r^2}$ using the radius (r) and mass (M) of earth with the gravitational constant G
# Thus, g is directly related to the mass and inversely proportional to the squared radius of te planet.
# We can observe this relation by looking at some graphs
```

```python
# Make a graph of (M/r^2) vs. gravity
x_var = np.array(df3['M/r^2'])
y_var = np.array(df3['gravity'])
plt.plot(x_var, y_var, 'bo-')
plt.title("M/r² vs. Acceleration of Gravity")
plt.xlabel("M/r² (kg/m²)")
plt.ylabel("Acceleration of Gravity (m/s²)")
plt.show()
# Slope should be G (6.674e-11)
G = (23.1-0.7) / (3.71348193e+11-1.03971942e+10)
print(f'Slope using first and last points: {G:.3e}') # 6.206e-11
print(f'Percent error from gravitational constant: {abs((G - 6.674e-11) / 6.674e-11 * 100):.2f}%') # 7.01%
print("We have confirmed Newton's law holds across real NASA data of planets.")
```
